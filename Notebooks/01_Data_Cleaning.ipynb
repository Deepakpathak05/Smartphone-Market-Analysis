#### G. Clean camera column
The camera information is stored as a single text string containing multiple details such as number of front and rear camera, primary rear camera and primary front camera specifications. Splitting this information enables more detailed analysis.

```python
df1['camera'].value_counts()
```

```python
def camera_extractor(text):
    if 'Quad' in text:
        return 4
    elif 'Triple' in text:
        return 3
    elif 'Dual' in text:
        return 2
    else:
        return 1
```

#### I. Clean os column

```python
df1['os'].value_counts()
```

```python
def os_extractor(text):
    if 'Android' in text:
        return 'android'
    elif 'iOS' in text:
        return 'ios'
    elif 'Harmony' in text or 'Hongmeng' in text:
        return 'other'
```
